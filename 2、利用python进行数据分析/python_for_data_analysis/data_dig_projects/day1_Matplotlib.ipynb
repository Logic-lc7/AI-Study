### 1、折线图

#### ①绘制折线图

```python
# 导入折线图包
import matplotlib.pyplot as plt
import random

# 设置中文
from pylab import mpl
mpl.rcParams["font.sans-serif"] = ["SimHei"]

# 生成数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]

# 创建画布
plt.figure(figsize=(15,4), dpi=80)  # 设置图像大小及画质

# 绘制图像
plt.plot(x, y_shanghai, color="r", linestyle="--", label="上海")  # 设置折线风格
plt.plot(x, y_beijing, color="b", label="北京")  # 绘制第2条折线

# 添加图例
plt.legend()

# 修改x，y轴刻度
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5], x_label[::5])
plt.yticks(range(0, 40, 5))

# 添加网格显示
plt.grid(linestyle="--", alpha=0.5)

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("中午11点到12点之间的温度变化")

# 保存图片
plt.savefig("two_city_temperature.png")

# 展示图片
plt.show()
```

#### ②绘制函数图像

```python
import numpy as np
import matplotlib.pyplot as plt

# 显示负号
import matplotlib 
matplotlib.rcParams['axes.unicode_minus'] =False

x = np.linspace(-10, 10, 1000)
y = np.sin(x)

plt.figure(figsize=(10, 4), dpi=80)
plt.plot(x, y)
plt.grid()
plt.show()
```
